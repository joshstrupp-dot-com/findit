```python
import pandas as pd
temp_df = pd.read_csv('zeroshot_analysis_results.csv')
print("Count of non-missing predicted labels:", temp_df['predicted_label'].notna().sum())
print("Count of 'very harmful' predicted labels:", (temp_df['predicted_label'] == 'Very Harmful').sum())


import matplotlib.pyplot as plt

# Count the frequency of each predicted label
label_counts = temp_df['predicted_label'].value_counts()

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(label_counts.index, label_counts.values)

# Customize the plot
plt.title('Distribution of Predicted Labels')
plt.xlabel('Predicted Label')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()

```
